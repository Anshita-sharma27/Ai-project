#numpy project

**employee salary**

```python
import numpy as np
```

```python
name= np.array(["anshita", "anchal", "devansh"," mohit", "khushi", "diksha"])

salaries = np.array([30000,40000,50000,60000,70000,80000])
```

```python
#find salary of each employ
for i in range(len(name)):
  print(name[i],salaries[i])
```

```python
average_salary=np.mean(salaries)
highest_salary=np.max(salaries)
lower_salary=np.min(salaries)
salary_range=highest_salary-lower_salary


below_avg=salaries[salaries<average_salary]
above_avg= salaries[salaries<average_salary]

print("average salary", average_salary)
print("higest salary", highest_salary)
print("lowest salary", lower_salary)
print("salary  range", salary_range)
print("\n employess below average salary:", below_avg)
print("\n employess above average salary:", above_avg)

```
